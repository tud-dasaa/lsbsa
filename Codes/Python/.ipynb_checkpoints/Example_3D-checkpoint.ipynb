## Least Squares B-Spline Approximation (LSBSA)
### Application of LSBSA to approximate <u>3D manifold</u> fitting problem

```python
# Load the necessary packages 
import numpy as np
from scipy import linalg
from matplotlib import pyplot as plt
from scipy.stats import norm
from scipy import signal
from scipy.stats.distributions import chi2
from scipy.interpolate import BSpline, PPoly
import scipy.io
```

```python
# Example on approximation of a known manifold f(u,v,t) using LSBSA
# import least-squares B-spline approximation (LSBSA) functions
import lsbsa_package as bs
# introduce m number of observations
# generate 10000 random positions in u-v plane between -2 and 2
m = 10000
u = 4 * (np.random.rand(m) - 0.5)
v = 4 * (np.random.rand(m) - 0.5)
t = 4 * (np.random.rand(m))
z = (1 + t / 20) * u * np.exp(-u ** 2 - v ** 2) + v * u * (t / 50) # % known 3D mathematical function f(u,v,t)
# observations y (without error) in the model y = Ax + e
y = z
# introduce knot intervals for u, v, and t
eps1 = 1e-5;
knotu = np.array([-2-eps1, -1.5, -1, -0.5, 0, 0.5, 1, 1.5, 2+eps1])
knotv = np.array([-2-eps1, -1.5, -1, -0.5, 0, 0.5, 1, 1.5, 2+eps1])
knott = np.array([0-eps1, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4+eps1])
# DEFINE the B-Spline degrees
# Bspline degree along u, v and t axes : 1:Linear 2:Quadratic 3:Cubic 4: Quartic, etc
degu = 3
degv = 3
degt = 3
# establish the design matrix in the model y=Ax+e
A = bs.A_matrix_3D(u, v, t, knotu, knotv, knott, degu, degv, degt)
m, n = A.shape
# apply the least squares method to estimate x, y, e and sigma
xhat, yhat, ehat, sigma = bs.lsfit(A,y) 
Sig = sigma    
# plot the residuals of the fit
plt.plot(ehat, 'o')
plt.xlabel('Observation ID')
plt.ylabel('Residual')
plt.box(on=True)
plt.grid(True)
plt.show()
# plot simulated data z = f(u,v,t)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(u, v, t, c=z, cmap='jet', s=4)
plt.colorbar(sc)
ax.set_xlabel('u-axis')
ax.set_ylabel('v-axis')
ax.set_zlabel('t-axis')
plt.grid(True)
plt.box(True)
plt.show()

# make a regular grid to predict f(u,v,t) values
U, V = np.meshgrid(np.arange(-2, 2.1, 0.1), np.arange(-2, 2.1, 0.1))
m1, n1 = U.shape
up = U.reshape(m1 * n1, 1)
vp = V.reshape(m1 * n1, 1)
t = np.array([0, 1, 2, 3])
for i in range(len(t)):
    # for any time instance t=0,1,2,3 predict f(u,v,t)
    T = np.ones((m1, n1)) * t[i]
    tp = T.reshape(m1 * n1, 1)
    Ap = bs.A_matrix_3D(up, vp, tp, knotu, knotv, knott, degu, degv, degt)
    yp = Ap @ xhat
    Yp = yp.reshape(m1, n1)
    # plot the predicted values f(u,v,t=0,1,2, or 3)
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(U, V, Yp, cmap='jet')
    plt.colorbar(surf)
    ax.set_xlabel('u-axis')
    ax.set_ylabel('v-axis')
    ax.set_zlabel('f(u,v,t)')
    ax.set_zlim(-0.6, 0.6)
    plt.grid(True)
    plt.box(True)
    plt.title('t = ' + str(t[i]))
    plt.show()

#plt.savefig('fig.pdf')
#plt.savefig('fig.tif', dpi=300)
```
